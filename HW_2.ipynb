```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
def root_mean_square_deviation(arr, perfect):
    return np.sqrt(np.mean(np.square(perfect - arr)))

graph1 = []
graph2 = []
for k in range(1, 100):
    arr1 = []
    arr2 = []
    for _ in range(100):
        arr1.append((np.mean(np.random.uniform(0, 1, 100) ** k) * (k + 1)) ** (1 / k))
        arr2.append((np.mean(np.random.exponential(1, 100) ** k) / np.math.factorial(k)) ** (1 / k))
    graph1.append(root_mean_square_deviation(np.array(arr1), 1))
    graph2.append(root_mean_square_deviation(np.array(arr2), 1))
plot.figure(figsize=(15, 10))
plot.xlabel("k")
plot.ylabel("standard deviation")
plot.title("uniform")
plot.plot(range(1, 100), graph1)

plot.figure(figsize=(15, 10))
plot.xlabel("k")
plot.ylabel("standard deviation")
plot.title("exponential")
plot.plot(range(1, 100), graph2)
plot.show()
```
