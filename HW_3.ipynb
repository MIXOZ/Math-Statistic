```python
import numpy as np
import matplotlib.pyplot as plt
from sortedcontainers import SortedSet
```

```python
def graph_diff_n_a(n):
    for alpha in range(1, 11, 2):
        arr = []
        for i in range(1000):
            tmp_arr = np.random.exponential(1/alpha, n)
            tmp_arr.sort()
            arr.append(tmp_arr)
        
        plt.plot(
            range(0, n),
            np.array(arr).mean(axis=0),
            label=alpha
        )
    plt.title("n = " + str(n))
    plt.legend()
    plt.xlabel("human's num")
    plt.ylabel("average waiting time")
    plt.show()
```

```python
for i in [10, 100, 500, 1000, 10000]:
    graph_diff_n_a(i)
```

```python
def graph_diff_n_b(n):
    for alpha in range(1, 11, 2):
        arr = []
        for k in range(100):
            tmp_arr = SortedSet(np.random.exponential(1/alpha, n))
            res = [0] * 1000
            for i in range(1000):
                res[i] = tmp_arr.pop(0)
                tmp_arr.add(np.random.exponential(1/alpha) + res[i])
            arr.append(res)
        plt.plot(range(0, 1000), np.array(arr).mean(axis=0), label="Alpha={}".format(alpha))
    plt.title("n = " + str(n))
    plt.legend()
    plt.xlabel("human's num")
    plt.ylabel("average waiting time")
    plt.show()
```

```python
for i in [10, 100, 500, 1000, 10000]:
    graph_diff_n_b(i)
```
